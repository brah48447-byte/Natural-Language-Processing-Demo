```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
texts = ['I love apples', 'I hate bananas']
vectorizer = CountVectorizer()#tokenize
X = vectorizer.fit_transform(texts)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```

```python
data = pd.read_csv('Feelings - Sheet1.csv')

X_train, X_test, y_train, y_test = train_test_split(data['Text'], data['Label'], test_size = 0.2)
```

```python
vectorizer = CountVectorizer()
X_train_counts = vectorizer.fit_transform(X_train)
X_test_counts = vectorizer.transform(X_test)

model = MultinomialNB()
```

```python
model.fit(X_train_counts, y_train)
prediction = model.predict(X_test_counts)

accuracy = accuracy_score(prediction, y_test)
print(prediction)
print(y_test)
print(accuracy)
```

```python
new_texts = ["I am happy", "This is terrible", "What a fun day!"]
labels = ["Positive", "Negative", "Positive"]
new_counts = vectorizer.transform(new_texts)
prediction_2 = model.predict(new_counts)

new_score = accuracy_score(prediction_2, labels)
print(prediction_2)
print(new_score)
```
